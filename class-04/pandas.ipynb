# Selecting data and put into new dataset

```python
df["Population_millions"] = df["Population"] / 1000000 #add a column to the dataframe
df["Population_millions"].head()
```

```python
df[df["Continent"] == "Asia"] #create a new subset based on condition
#options: df["Continent"].isin(["Asia", "Africa"]) to check if term in list
```

```python
asiancountries = df[df["Continent"] == "Asia"] #can store it in new variable
asiancountries.head()
```

```python
asiancountries_highgdp = df[(df['Continent'] == 'Asia') & (df['GDP_per_capita'] > df['GDP_per_capita'].mean())]
asiancountries_highgdp.head() #add several conditions - similar to sql-WHERE
```

```python
df1 = df[["Population", "Continent"]] #Put several columns into a new dataset
df1.head()
```
